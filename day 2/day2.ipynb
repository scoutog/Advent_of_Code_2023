```python
import pandas as pd
import re
```

```python
df = pd.read_csv("DAY2.csv")
# df = df.iloc[0:1,]
df[['GameNum','Outputs']] = df['INPUT'].str.split(':',expand=True)
df = df[['GameNum','Outputs']]
# df = df.iloc[0:1]
df["FEASIBILITY"] = "NO"
df
```

```python
colors = ['green', 'blue', 'red']

red_cap = 12
green_cap = 13
blue_cap = 14

for index, row in df.iterrows():
    red_ls = []
    green_ls = []
    blue_ls = []
    string = row['Outputs']
    by_game = string.split("; ")
    for game in by_game:
        outputs = game.split(",")
#         print(game)
        for output in outputs:
            for color in colors:
                if color in output:
#                     print(output)
                    if (color == 'green'):
                        green_ls.extend(re.findall(r'\d+', output))
                        
                    if (color == 'blue'):
                        blue_ls.extend(re.findall(r'\d+', output))
                        
                    if (color == 'red'):
                        red_ls.extend(re.findall(r'\d+', output))

    green_ls = [eval(i) for i in green_ls]
    blue_ls = [eval(i) for i in blue_ls]
    red_ls = [eval(i) for i in red_ls]
    
    if (max(green_ls)<=green_cap) & (max(blue_ls)<=blue_cap) & (max(red_ls)<=red_cap):
        df.iloc[index, 2] = 'YES'
```
